```python
from core.reader import read_local_data
from config import DB_ROOT_PATH
import json
```

```python
def get_all_trading_pairs(dbinfo_path,exchange_id):
    file_path = dbinfo_path
    with open(file_path, 'r', encoding='utf-8') as f:
        db_info = json.load(f)
    pairs_list = list(db_info[exchange_id].keys())
    return pairs_list
```

```python
pairs_list = get_all_trading_pairs(
    r"F:\MyCryptoTrading\CryptoTradingSystem_allin1\Cryptodataklines_featherdb\infomation\dbinfo.json",
    "binance"
    )
```

```python
pairs_list = [_  for _ in pairs_list if _[-4:]=="USDT"]
```

```python
kline_data = read_local_data(
    db_root_path=DB_ROOT_PATH,
    exchange_id="binance",
    interval="1h",
    kline_type="swap",
    num_workers=16,
    symbols=pairs_list
)
```
