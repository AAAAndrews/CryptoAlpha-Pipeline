```python
from deap import gp
count = 0
success_to_transform = []
for i in df['expression_']:
    try:
        if str(gp.PrimitiveTree.from_string(i, pset)).replace(" ","")!=i.replace(" ",""):
            print(i,"\n",gp.PrimitiveTree.from_string(i, pset))
            success_to_transform.append(False)
        elif str(gp.PrimitiveTree.from_string(i, pset)).replace(" ","")==i.replace(" ",""):
            success_to_transform.append(True)
    except:
        print("错误",i)
        success_to_transform.append(False)
        count+=1
print("总错误数:",count)
```
