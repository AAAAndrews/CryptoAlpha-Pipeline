```python
import pandas as pd
import joblib
```

```python
alpha191 = pd.read_excel("alphas.xlsx",sheet_name="alpha191")
alpha191 = alpha191[alpha191["是否为三元表达式"]=="否"].reset_index(drop=True)
str_list = []
for i in range(0,len(alpha191),5):
    if i +5<= len(alpha191):
        expr = alpha191.loc[i:i+5, "因子公式"]
    elif i < len(alpha191):
        expr = alpha191.loc[i:len(alpha191)-1, "因子公式"]
    else:
        break
    numeric_str = ";".join(expr.tolist())
    str_list.append(numeric_str)
joblib.dump(str_list, "numericalpha_str_alpha191.pkl")
```

```python
df = pd.DataFrame(joblib.load('./wq_formatted_alpha191.pkl'))

expr_cols = [col for col in df.columns if 'expression' in col or 'detail' in col]

# 创建合并列
df['expression_'] = df[expr_cols].apply(
    lambda row: ' | '.join(filter(lambda x: pd.notna(x) and 'Missing' not in str(x), row)), 
    axis=1
)
```
